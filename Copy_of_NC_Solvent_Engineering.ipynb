```python
def Ra(sol, target):
    return np.sqrt(
        4*(sol[0]-target["dd"])**2 +
          (sol[1]-target["dp"])**2 +
          (sol[2]-target["dh"])**2
    )
```

```python
import itertools

# our Target Solute
imidazole = {
    "dd": 19.7,
    "dp": 15.6,
    "dh": 11.2,
    "R0": 6.0
}

results = []

fractions = np.linspace(0.1, 0.9, 9)

for (n1, s1), (n2, s2) in itertools.combinations(solvents.items(), 2):
    for f in fractions:
        mix = f*s1 + (1-f)*s2
        r = Ra(mix, imidazole)
        label = f"{n1}:{n2} ({f:.1f}:{1-f:.1f})"
        results.append((label, r))
```

```python
results.sort(key=lambda x: x[1])

for name, r in results[:100]:
    print(f"{name:40s}  Ra = {r:.2f}")
```
